## 작업형 1유형 대비 - 필터

- 다음은 2000년대 중반 관측된 ufo 데이터이다.

- 다음 중 `State`가 'CA'이고, `Shape Reported` 가 'Light'인 행의 `Colors Reported`을 'Red'로 바꾸시오. (단 값이 있는 경우 제외한다.)

- 이 후 `State`가 'CA'이며, `Shape Reported`가 'Red'인 행의 개수를 추출하시오.

> 값은 `print()`함수를 통해 출력하시오 (숫자만)

```python
import pandas as pd
```

```python
df = pd.read_csv('ufo.csv')
```

```python
df.head(10)
```

```python
df[(df['Shape Reported'] == 'LIGHT') & (df['State'] == 'CA')]
```

```python
df[(df['Shape Reported'] == 'LIGHT') & (df['State'] == 'CA') & (df['Colors Reported'].isna())]
```

```python
df['Colors Reported'][(df['Shape Reported'] == 'LIGHT') & (df['State'] == 'CA') & (df['Colors Reported'].isna())] = 'RED'
```

```python
df[(df['Shape Reported'] == 'LIGHT') & (df['State'] == 'CA') & (df['Colors Reported'].isna())]
```

```python
df[(df['State'] == 'CA') & (df['Colors Reported'] == 'RED')]
```
