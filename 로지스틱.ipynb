```python
import os
import pandas as pd

x_train=pd.read_csv('X_train.csv',encoding='euc-kr')
y_train=pd.read_csv('y_train.csv',encoding='euc-kr')
x_test=pd.read_csv('X_test.csv',encoding='euc-kr')

x_train
```

## 1) 결측 데이터 처리

```python
x_train.info()
```

### 1-1. 결측치 제거




```python
x_train = x_train.fillna(0)
x_test = x_test.fillna(0)
x_train
```

### 1-2. 범주형변수 변환 (인코딩)

```python
## Pandas의 factorize : sklearn의 라벨 인코딩과 유사 ( 빈도 순으로 인코딩 )

# pd.factorize(x_train["주구매상품"]) 의 결과는 factorize의 값([0])과 인덱스([1])가 나옴
# .reshape(-1,1)의 경우 행은 가변적, 열은 1행으로 도출 / 해당 분석에는 굳이 쓸 필요는 없는듯 (유무 상관없이 결과 같음)
x_train.loc[:, "주구매상품"] = pd.factorize(x_train["주구매상품"])[0].reshape(-1,1)
x_train.loc[:, "주구매지점"] = pd.factorize(x_train["주구매지점"])[0].reshape(-1,1)


x_test.loc[:, "주구매상품"] = pd.factorize(x_test["주구매상품"])[0].reshape(-1,1)
x_test.loc[:, "주구매지점"] = pd.factorize(x_test["주구매지점"])[0].reshape(-1,1)

# x_train['gender'] = y_train['gender'] # label을 트레이닝 데이터에 합침, 데이터 분리를 위해서.
x_train.head()
```

```python
# cust_id 열 제거 (아이디 삭제)
x_train = x_train.iloc[:, 1:]
y_train = y_train['gender']

x_test = x_test.iloc[:,1:] # 3500 ~ 5981
```

## 2) 데이터 스케일링

> * 이것은 엔지니어의 약속으로 변환을 하면 안된다. (비추천)

```python
# 개념은 스케일러를 train에 맞춰두고 test에 그대로 진행
# 각자 진행하면 오류발생 (엄청난 영향을 준다.)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
# x_test_scaled[0]
```

## 3) 모델 만들기

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

## 4) 훈련

```python
model.fit(x_train_scaled,y_train)
```

```python
x_test
```

## 5) 예측 (해당 레이블로 분류될 확률 값을 알고 싶다)

```python
predict_proba = model.predict_proba(x_test_scaled)

# 앞은 0 뒤는 1일 확률
predict_proba
```

```python
# predict vs predict_proba : 하나의 샘플 vs 각각 모든 샘플
submit = pd.DataFrame(predict_proba[:,1],columns=['gender'])
submit
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
def train_and_test(model):
    model.fit(x_train_scaled,y_train) # 훈련 세트(x_train)를 타겟값(y_train)에 대하여 학습
    prediction = model.predict(x_test_scaled) # 테스트(x_test) 데이터 세트에 대해 타겟값 예측
    accuracy = round(model.score(x_train_scaled,y_train) * 100, 2) # 10번 시행시 평균 정확도
    print("정확도 : ", accuracy, "%")
    return prediction
```

```python
log_pred = train_and_test(LogisticRegression())
```

```python
svm_pred = train_and_test(SVC())
```
