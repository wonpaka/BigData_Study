### Join & groupby

> Q1. 다음은 2000년 8월말에 발견된 UFO 데이터 이다. 학자들은 UFO 발견 당시 날씨를 기온을 알고 싶어한다.
>
>    미국 전역의 기상 데이터를 통해 날짜별 최고 기온의 평균을 산출하고 UFO 발견날짜에 맞춰 데이터를 생성 한 뒤 저장하여라.
>
>    (소수 둘째 자리에서 반올림)
    
    - weather.csv, ufo.csv

```python
import pandas as pd
```

```python
ufo = pd.read_csv('ufo.csv')
weather = pd.read_csv('weather.csv')
```

```python
ufo.head()
```

```python
weather.head()
```

```python
ufo['Date'] = ufo['Time'].str[:4]
```

```python
weather['Date'] = weather['day'].str[6:10].str.replace('-','/')
```

```python
weather_gb = weather.groupby('Date').mean().reset_index()
weather_gb['highest'] = weather_gb['highest'].round(2)
weather_gb = weather_gb[['Date', 'highest']]
weather_gb.head()
```

```python
answer = pd.merge(ufo,weather_gb,on='Date',how='left')
answer.head(20)
```

```python
answer.to_csv('Q1.csv')
```
